# Data Assessing and Cleaning
1. create subset based on requirements
2. drop duplicates
3. convert datatypes

```python
# Selecting columns
columns = ["ListingKey",
           "ListingNumber",
           "ListingCreationDate",
           "ListingCategory (numeric)",
           "LoanStatus",
           "ClosedDate",
           "Term",
           "BorrowerAPR",
           "BorrowerRate",
           "ProsperRating (Alpha)",
           "ProsperScore",
           "Occupation",
           "EmploymentStatus",
           "CreditScoreRangeLower",
           "CreditScoreRangeUpper",
           "DateCreditPulled",
           "DebtToIncomeRatio",
           "StatedMonthlyIncome",
           "LoanOriginalAmount",
           "MonthlyLoanPayment",
           "AmountDelinquent",
           "IncomeRange"]
loan = df[columns]
```

```python
def show_data_summary(df: pd.DataFrame):
    print ("Shape: {}".format(df.shape))
    print (df.info())
    print (df.head())
```

```python
show_data_summary(loan)
```

```python
# Check if data has duplicates
loan.duplicated().sum()
```

```python
# Changing data type for colums "ListingCreationDate","ClosedDate","DateCreditPulled".
datetime_change = ["ListingCreationDate","ClosedDate","DateCreditPulled"]
loan[datetime_change] = loan[datetime_change].apply(pd.to_datetime)
```

```python
# keeping on date and removing time from column data
for column_name in datetime_change:
    loan[column_name] = pd.to_datetime(df[column_name]).dt.date
```

```python
loan.info()
```

```python
# Fill missing data
loan['DebtToIncomeRatio'] = loan['DebtToIncomeRatio'].fillna(loan['DebtToIncomeRatio'].mean())
```

```python
# Check for null values after filling missing data
loan['DebtToIncomeRatio'].isnull().sum()
```

```python
# Get rows that are not null from prosper score column 
loan = loan[loan['ProsperScore'].notnull()]
```

## Univariate Exploration

```python
def draw_hist(df: pd.DataFrame, x: str, bins = None, title = ''):
    """
    To draw histogram for univariate data
    """
    if bins is not None:
        plt.hist(data = df, x = x,bins = bins)
    else:
        plt.hist(data= df, x = x)
    plt.xlabel(x)
    plt.title(title);
```

```python
# distribution of loan original amount.
bins = np.arange(0, 40000, 5000)
    
draw_hist(loan, 'LoanOriginalAmount', bins, title = 'Distribution of loan original data')   
```

By seeing the visulaization we can say that, there are more number of loans with the amount of below 5000K and small number of loans in the amount from 20K to 35K.

```python
#distribution of AmountDelinquent
plt.figure(figsize=[10,6])
bins = np.arange(0,100000, 5000)

draw_hist(loan, 'AmountDelinquent', bins, title = 'Distribution of amount deliquent')
```

## Bivariate Exploration

Let us see how occupation and employment status of the borrower effects each their monthly loan payments:-

```python
def draw_barplot(df, x, y, color = sb.color_palette()[0], title = ''):
    """
    To draw barplot for bivariate data
    """
    sb.barplot(data = df, x = x, y = y, color = color)
    plt.ylabel(y)
    plt.xlabel(x)
    plt.title(title)
```

```python
plt.figure(figsize=[6,30])

plt.subplot(2,1,1)
draw_barplot(loan, 'MonthlyLoanPayment', 'Occupation', title = 'Effects of occupation on monthly loan payments')
plt.xticks(rotation=90)

plt.subplot(2,1,2)
draw_barplot(loan, 'MonthlyLoanPayment','EmploymentStatus', title = 'Effects of emplyment status on monthly loan payments')
plt.xticks(rotation=90);
```

The top 3 occupations have high monthly loan payment are Judge, Doctor,  and Pharmacist. The plot is consistent with salaries of each occupation.
we can say this by seeing plot 2 also employed have high monthly loan payment.

Let us look into relationship of amount deliquent and employment status:-

The borrowers with other employment status have more delinquencies. It means borrowers with out proper income have more delinquencies.

The borrower APR is negative correlated with loan amount. It means decrease in APR with the increase of loan amount.

## Multivariate Exploration

Let us see how loan original amount and term of the loan effect the prosper rating:-

```python
credit_rating = ['AA','A','B','C','D','E','HR']
sb.pointplot(data=loan, x = 'ProsperRating (Alpha)', y = 'LoanOriginalAmount', hue = 'Term', order = credit_rating);
plt.xlabel('ProsperRating (Alpha)')
plt.ylabel('Loan Original Amount')
plt.title('Effects of loan original amount and term of the loan on prosper rating');
```

By seeing the visualization, we can say that the loan amount for all terms increases with good prosper rating and also there is increase in term for larger loan amount.

let us see the relationship between credit score ,monthly income and prosper rating:-

```python
credit_rating = ['AA','A','B','C','D','E','HR']
plt.figure(figsize = [10, 6])
plot = sb.stripplot(data=loan.query('StatedMonthlyIncome<StatedMonthlyIncome.quantile(0.995)'),
                                    x='CreditScoreRangeLower',
                                    y='StatedMonthlyIncome',
                                    hue='ProsperRating (Alpha)',
                                    hue_order=credit_rating,
                                    size=2, jitter=0.35, dodge=True)
plt.title('Relationship between credit score ,monthly income and prosper rating');
```

The most of the borrowes have good credit score when their income is high. only few borrower have 840 above credit score and AA rating.The borrowers with credit score above 660 have all types of rating except AA.
