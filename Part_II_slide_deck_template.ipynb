## Dataset Overview
The data set contains 84853 loan with 22 columns information like BorrowerAPR, ProsperScore, MonthlyLoanPayment, credit score, StatedMonthlyIncome and so on.

I created subset of the columns to 22 from 81 columns.
looked for duplicate rows.
converted datatypes of ListingCreationDate, ClosedDate, DateCreditPulled to datatime and seperated time from columns.
Deleted the rows with out prosper score.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df = pd.read_csv('prosperLoanData.csv')
```

```python
columns = ["ListingKey",
           "ListingNumber",
           "ListingCreationDate",
           "ListingCategory (numeric)",
           "LoanStatus",
           "ClosedDate",
           "Term",
           "BorrowerAPR",
           "BorrowerRate",
           "ProsperRating (Alpha)",
           "ProsperScore",
           "Occupation",
           "EmploymentStatus",
           "CreditScoreRangeLower",
           "CreditScoreRangeUpper",
           "DateCreditPulled",
           "DebtToIncomeRatio",
           "StatedMonthlyIncome",
           "LoanOriginalAmount",
           "MonthlyLoanPayment",
           "AmountDelinquent",
           "IncomeRange"]
loan = df[columns]
```

```python
datetime_change = ["ListingCreationDate","ClosedDate","DateCreditPulled"]
loan[datetime_change] = loan[datetime_change].apply(pd.to_datetime)
```

```python
for column_name in datetime_change:
    loan[column_name] = pd.to_datetime(df[column_name]).dt.date
```

```python
loan['DebtToIncomeRatio'] = loan['DebtToIncomeRatio'].fillna(loan['DebtToIncomeRatio'].mean())
```

```python
loan = loan[loan['ProsperScore'].notnull()]
```

# Distribution of Borrower APR
In the distribution of Borrower APR, there are more loans with 3.5 and 3.6 APRate and very few loans have APR greater than 4.3

```python
plt.figure(figsize=[10,6])
bins =  np.arange(0, loan.BorrowerAPR.max()+0.02, 0.01)
plt.hist(loan['BorrowerAPR'], bins=bins)
plt.title('Distribution of Borrower APR')
plt.xlabel('Borrower APR')
plt.ylabel('count');
```

# Distribution of loan original amount
There are more number of loans with the amount of below 5000K and small number of loans in the amount from 20K to 35K.

```python
plt.hist(loan['LoanOriginalAmount'])
plt.title('Distribution of loan original amount')
plt.xlabel('loan original amount')
plt.ylabel('count');
```

# Loan original amount Vs. Borrower APR

By seeing the data of the borrower APR and loan original amount. They have negative relationship. It means large amount of loans have low APR and smaller loans have large APR.

```python
sb.regplot(data=loan, x='LoanOriginalAmount',y='BorrowerAPR',scatter_kws={'alpha':0.01})
plt.title('Borrower APR vs. loan original amount')
plt.xlabel('Loan Original Amount')
plt.ylabel('Borrower APR');
```

# Borrower APR effect by prosper rating and term

The borrower APR increases when the term increses for B, A and AA rating. The borrower APR decreses when term increases for D, E and C.


```python
fig = plt.figure(figsize = [10,6])
ax = sb.pointplot(data = loan, x = 'ProsperRating (Alpha)', y = 'BorrowerAPR', hue = 'Term',linestyles = '',dodge = 0.4)
plt.title('Borrower APR across rating and term')
plt.ylabel('Mean Borrower APR')
plt.xlabel('Prosper Rating');
```

# effects of monthly loan payment by occupation and employment status

The top 3 occupations have high monthly loan payment are Judge, Doctor,  and Pharmacist. The plot is consistent with salaries of each occupation.
we can say this by seeing plot 2 also employed have high monthly loan payment.

```python
plt.figure(figsize=[6,30])
color = sb.color_palette()[0]

def draw_barplot(x, y):
    sb.barplot(data=loan, x=x,y=y,color=color)
    plt.ylabel(y)
    plt.xlabel(x)
    plt.xticks(rotation=90);

plt.title('loan payments across the occupation')    
draw_barplot('MonthlyLoanPayment', 'Occupation')
```

```python
plt.title('loan payments across the employment status') 
draw_barplot('MonthlyLoanPayment','EmploymentStatus')
```

# Monthly loan paymeny vs. Borrower rate

The borrower rate and monthly payments shows negative correlation. when loan payment amount is large when rate is low.

```python
sb.regplot(data=loan,x='MonthlyLoanPayment',y='BorrowerRate',scatter_kws={'alpha':0.01})
plt.title('Monthly loan paymeny vs. Borrower rate')
plt.xlabel('Monthly Loan Payment')
plt.ylabel('Borrower Rate');
```
